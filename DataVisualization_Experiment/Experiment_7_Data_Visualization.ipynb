# Experiment 7: Data Visualization using Matplotlib and Seaborn
**Aim:** To visualize data using Matplotlib and Seaborn libraries in Python and perform univariate and bivariate analysis.

### Step 5: Import Required Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Step 6: Load Dataset
We use Seaborn's built-in `diamonds` dataset.

```
df = sns.load_dataset("diamonds")
df.head()
```

### Step 7: Univariate Analysis
#### 7.1 Histogram

```
sns.histplot(df["price"], kde=True)
plt.title("Distribution of Diamond Prices")
plt.show()
```

#### 7.2 Boxplot - Univariate

```
sns.boxplot(y=df["price"])
plt.title("Boxplot of Diamond Prices")
plt.show()
```

#### 7.3 Countplot

```
sns.countplot(x="cut", data=df)
plt.title("Count of Diamond Cuts")
plt.show()
```

#### 7.4 Scatterplot

```
sns.scatterplot(x="carat", y="price", data=df)
plt.title("Carat vs Price")
plt.show()
```

#### 7.5 Regplot

```
sns.regplot(x="carat", y="price", data=df.sample(1000, random_state=42), scatter_kws={'alpha':0.5})
plt.title("Regression: Carat vs Price")
plt.show()
```

### Step 8: Bivariate Analysis
#### 8.1 Boxplot Bivariant

```
sns.boxplot(x="cut", y="price", data=df)
plt.title("Price by Cut")
plt.show()
```

#### 8.2 Violinplot

```
sns.violinplot(x="cut", y="price", data=df)
plt.title("Price Distribution by Cut")
plt.show()
```

#### 8.3 Histplot-Seaborn

```
sns.histplot(df['price'], bins=10, kde=True)
plt.title("Histogram of Diamond Price using Seaborn")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

#### 8.4 Lineplot

```
sns.lineplot(x='carat', y='price', data=df)
plt.title("Line Plot of Carat vs Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```
