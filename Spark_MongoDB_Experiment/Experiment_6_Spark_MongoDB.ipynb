# Experiment 6: Connecting Spark with MongoDB
**Aim:** To install Apache Spark and connect it with MongoDB to perform Big Data processing.

### Step 6 & 7: Initialize PySpark in Jupyter
We use `findspark` to link with the PySpark libraries.

```
import findspark
findspark.init()
```

### Step 8: Create Spark Session with MongoDB Connector
We initialize the Spark session and pass the MongoDB Spark Connector as a configuration package.

```
from pyspark.sql import SparkSession

try:
    spark = SparkSession.builder \
        .appName("SparkMongoDemo") \
        .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:3.0.1") \
        .config("spark.mongodb.output.uri", "mongodb://127.0.0.1:27017/sparkdb.employee") \
        .config("spark.mongodb.input.uri", "mongodb://127.0.0.1:27017/sparkdb.employee") \
        .config("spark.ui.port", "4050") \
        .getOrCreate()
    print("Spark connected successfully")
except Exception as e:
    print(f"Failed to connect to Spark: {e}")
```

### Step 9: Create DataFrame and Insert Data
Create a subset of mock structured employee data and insert it seamlessly into our NoSQL database natively using DataFrame writes.

```
data = [
    (1, "Ravi", "IT", 40000),
    (2, "Sita", "HR", 35000),
    (3, "Kiran", "Finance", 48000)
]
columns = ["empid", "name", "dept", "salary"]

df = spark.createDataFrame(data, columns)
df.show()
```

Write DataFrame directly back to MongoDB using `df.write.format("mongo")`.

```
try:
    df.write.format("mongo").mode("append").save()
    print("Data successfully inserted into MongoDB")
except Exception as e:
    print(f"Error: {e}")
```

### Verification: Read Data Back from MongoDB

```
df_read = spark.read.format("mongo").load()
df_read.show()
```

### Cleanup

```
spark.stop()
```
