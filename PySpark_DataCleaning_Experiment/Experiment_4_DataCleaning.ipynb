# Experiment 4: Data Cleaning and Transformation using PySpark DataFrames
**Aim:** To Implement Data cleaning and transformation using PySpark DataFrames.

### Step 6: Initialize PySpark in Jupyter
We use `findspark` to automatically link the Python environment with the PySpark libraries.

```
import findspark
findspark.init()
```

### Step 7: Create SparkSession
Initialize the Spark engine and UI port.

```
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("DataCleaningLab") \
    .config("spark.ui.port", "4050") \
    .getOrCreate()

spark
```

### Step 8: Run Data Cleaning Program
Create the initial DataFrame containing mock values, specifically including `None` elements to simulate missing data.

```
data = [
    (1, "AJITH Goud", 25, "Bangalore", 50000),
    (2, "JITHENDRA", None, "Hyderabad", 45000),
    (3, "HESHWANTH", 28, None, None),
    (4, "VISHNU", 35, "Chennai", 70000),
    (5, "MAHESH", 40, None, 40000)
]

columns = ["id", "name", "age", "city", "salary"]

df = spark.createDataFrame(data, columns)
df.show()
```

### Step 9: Perform Data Cleaning
Use `fillna` to dynamically clean up missing inputs across different columns, swapping `None` for integer `0` or string `'Unknown'`.

```
df_clean = df.fillna({
    "age": 0,
    "city": "Unknown",
    "salary": 0
})
df_clean.show()
```

### Step 10: Data Transformation Operations
#### 11.1 Select Specific Columns

```
df_select = df_clean.select("name", "salary")
df_select.show()
```

#### 11.2 Filter Rows (Salary > 40000)

```
df_filter = df_clean.filter(df_clean.salary > 40000)
df_filter.show()
```

#### 11.3 Create New Column (Annual Salary)
Apply an arithmetic multiplier to the `salary` column to generate a new feature column.

```
from pyspark.sql.functions import col

df_new = df_clean.withColumn("annual_salary", col("salary") * 12)
df_new.show()
```

#### 11.4 Rename Column

```
df_rename = df_new.withColumnRenamed("name", "employee_name")
df_rename.show()
```

#### 11.5 Sort Data (Descending Salary)

```
df_sort = df_rename.orderBy(df_rename.salary.desc())
df_sort.show()
```

#### 11.6 Group By and Aggregation (City-wise Average Salary)

```
from pyspark.sql.functions import avg

df_group = df_rename.groupBy("city").agg(avg("salary").alias("avg_salary"))
df_group.show()
```

### Step 11: Stop Spark Session (Important)
Always stop the Spark Session to gracefully release the Hadoop JVM resources.

```
spark.stop()
```
